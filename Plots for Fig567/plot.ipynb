## Fig 5.a

```python
arr = np.array([[4.683, 5.738, 5.2962, 5.45775], 
                [4.0894, 5.963, 5.0786, 10.4931],
                [0., 3.2433, 3.464, 3.9132],
                [0., 3.2459, 3.4713, 3.6151],
                [0., 2.9477, 3.6082, 6.2095],
                [0., 2.93175, 3.83258, 4.16214]
      ])

df = pd.DataFrame(arr)
df.index = ["Exp.1", "Exp.2", "Exp.3", "Exp.4", "Exp.5", "Exp.6"]
df.columns = ["IP", "QCS-DNN", "QCS-MAB", "Epsilon Greedy"]
# df.plot(kind='bar', width=0.8, alpha=0.7, color=["tab:blue", "tab:orange", "tab:green", "tab:red"])
df.plot(kind='bar', width=0.8, alpha=0.7, color=["#e7298a", "#1b9e77", "#d95f02", "#7570b3"])
plt.xticks(rotation=0)
plt.xlabel('Experiments', size=16)
plt.ylabel("Avg. Latency (sec)", size=16)
plt.xticks(fontsize=14)
plt.yticks(np.arange(0, 12.1, 2), fontsize=14)
plt.legend(loc='upper right', framealpha=0.3, fontsize=10)

plt.savefig("figs_py/avg_latency_exp.pdf", bbox_inches='tight')
plt.show()
```

```python
df
```

```python
val = list(df["IP"].values) + list(df["QCS-DNN"].values) + list(df["QCS-MAB"].values) + list(df["Epsilon Greedy"].values)
exp = list(df.index) * len(df.columns)
alg = ["IP"] * len(df.index) + ["QCS-DNN"] * len(df.index) + ["QCS-MAB"] * len(df.index) + ["Epsilon Greedy"] * len(df.index)
new_df = {"Avg. Latency (sec)": val, "Exp": exp, "Algorithms":alg}
new_df = pd.DataFrame(new_df)
new_df
```

```python
sns.histplot(data=new_df, x="Avg. Latency (sec)", hue="Algorithms", multiple="stack")
plt.xlabel('Avg. Latency (sec)', size=16)
plt.ylabel('Count', size=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.savefig("figs_py/avg_latency_exp.pdf", bbox_inches='tight')
plt.show()
```

```python
sns.kdeplot(data=new_df, x="Avg. Latency (sec)", hue="Algorithms", multiple="stack")
plt.xlabel('Avg. Latency (sec)', size=16)
plt.ylabel('Density', size=16)
plt.xticks(fontsize=14)
plt.yticks(np.arange(0, 0.251, 0.05), fontsize=14)
# plt.savefig("figs_py/avg_latency_exp.pdf", bbox_inches='tight')
plt.show()
```

```python
# Plot the distribution of clarity ratings, conditional on carat
sns.displot(
    data=new_df,
    x="Avg. Latency (sec)", hue="Algorithms",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
)
plt.xlabel('Avg. Latency (sec)', size=16)
plt.ylabel('Density', size=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.savefig("figs_py/avg_latency_exp.pdf", bbox_inches='tight')
plt.show()
```
